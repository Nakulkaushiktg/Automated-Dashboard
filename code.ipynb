**Total and Rank of Advance Screening**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data_path = 'Advance screening.csv'
data = pd.read_csv(data_path, encoding='latin1')  # Use appropriate encoding

# Clean and Prepare Data
data = data[['Name (Pls enter the same name as you did in preliminary screening- we have an automation in place, so you\'d have to enter the same name)',
             'Role Applied to', 'Total', 'Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10']]
data.columns = ['Name', 'Role', 'Total', 'Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10']

# Calculate Rankings
data['Rank'] = data['Total'].rank(method='max', ascending=False)
data = data.sort_values(by='Total', ascending=False)

# Calculate Average Scores by Question
average_scores = data[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10']].mean()

# Plot Leaderboard Table
print("\nLeaderboard:")
print(data[['Rank', 'Name', 'Role', 'Total']].sort_values(by='Rank').reset_index(drop=True))

# Plot Average Scores Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=average_scores.index, y=average_scores.values, palette='viridis')
plt.title('Average Scores by Question')
plt.ylabel('Average Score')
plt.xlabel('Question')
plt.xticks(rotation=45)
plt.show()

# Plot Heatmap of Scores
plt.figure(figsize=(12, 8))
sns.heatmap(data[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10']],
            cmap='coolwarm', annot=False, linewidths=0.5)
plt.title('Heatmap of Scores by Participant')
plt.xlabel('Question')
plt.ylabel('Participant Index')
plt.show()

# Plot Top Performers Line Chart
top_5 = data.nlargest(5, 'Total')[['Name', 'Total']]
plt.figure(figsize=(10, 6))
sns.lineplot(data=top_5, x='Name', y='Total', marker='o')
plt.title('Top 5 Performers by Total Score')
plt.ylabel('Total Score')
plt.xlabel('Name')
plt.xticks(rotation=45)
plt.show()

# Save Processed Data
processed_data_path = 'Advance Output.csv'
data.to_csv(processed_data_path, index=False)

# Provide the link to the saved leaderboard file
print(f"The leaderboard has been saved and can be accessed at: {processed_data_path}")
```

**Total and Rank of Basic Screening**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data_path = 'Basic Screening.csv'
data = pd.read_csv(data_path, encoding='latin1')  # Use appropriate encoding

# Clean and Prepare Data
data = data[['Name', 'Total', 'Q1', 'Q2', 'Q3', 'Q4', 'Q5']]
data.columns = ['Name', 'Total', 'Q1', 'Q2', 'Q3', 'Q4', 'Q5']

# Calculate Rankings
data['Rank'] = data['Total'].rank(method='max', ascending=False)
data = data.sort_values(by='Total', ascending=False)

# Calculate Average Scores by Question
average_scores = data[['Q1', 'Q2', 'Q3', 'Q4', 'Q5']].mean()

# Plot Leaderboard Table
print("\nLeaderboard:")
print(data[['Rank', 'Name', 'Total']].sort_values(by='Rank').reset_index(drop=True))

# Plot Average Scores Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=average_scores.index, y=average_scores.values, palette='viridis')
plt.title('Average Scores by Question')
plt.ylabel('Average Score')
plt.xlabel('Question')
plt.xticks(rotation=45)
plt.show()

# Plot Heatmap of Scores
plt.figure(figsize=(12, 8))
sns.heatmap(data[['Q1', 'Q2', 'Q3', 'Q4', 'Q5']], cmap='coolwarm', annot=False, linewidths=0.5)
plt.title('Heatmap of Scores by Participant')
plt.xlabel('Question')
plt.ylabel('Participant Index')
plt.show()

# Plot Top Performers Line Chart
top_5 = data.nlargest(5, 'Total')[['Name', 'Total']]
plt.figure(figsize=(10, 6))
sns.lineplot(data=top_5, x='Name', y='Total', marker='o')
plt.title('Top 5 Performers by Total Score')
plt.ylabel('Total Score')
plt.xlabel('Name')
plt.xticks(rotation=45)
plt.show()

# Save Processed Data
processed_data_path = 'Basic Output.csv'
data.to_csv(processed_data_path, index=False)

# Provide the link to the saved leaderboard file
print(f"The leaderboard has been saved and can be accessed at: {processed_data_path}")
```

```python
import pandas as pd
dataset1 = pd.read_csv("/content/Advance Output.csv")
dataset2 = pd.read_csv("/content/Basic Output.csv")
import pandas as pd


# Calculate the totals for Advance and Basic
dataset1['Advance'] = dataset1.filter(like='Q').sum(axis=1)
dataset2['Basic'] = dataset2.filter(like='Q').sum(axis=1)

# Merge datasets on Name
merged_df = pd.merge(dataset1[['Name', 'Advance']], dataset2[['Name', 'Basic']], on='Name', how='outer')

# Fill NaN values with 0
merged_df.fillna(0, inplace=True)

# Calculate the combined Total and Percentage
merged_df['Total'] = merged_df['Advance'] + merged_df['Basic']
merged_df['Percentage'] = (merged_df['Total'] / merged_df['Total'].max()) * 100

# Save the processed dataset to a new CSV file
merged_df.to_csv('Final LeaderBoard.csv', index=False)

print("The processed dataset has been saved as 'Final Leaderboard.csv'.")
```

**Graphs of my LeaderBoard**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the final leaderboard data
final_leaderboard = pd.read_csv('Final LeaderBoard.csv')

# Create a bar plot of the total scores
plt.figure(figsize=(10, 6))
sns.barplot(x='Name', y='Total', data=final_leaderboard)
plt.title('Total Scores for All Participants')
plt.xlabel('Participant Name')
plt.ylabel('Total Score')
plt.xticks(rotation=45)
plt.show()


# Create a scatter plot of Advance vs. Basic scores
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Advance', y='Basic', data=final_leaderboard)
plt.title('Advance vs. Basic Scores')
plt.xlabel('Advance Score')
plt.ylabel('Basic Score')
plt.show()


# Create a histogram of the percentage scores
plt.figure(figsize=(10, 6))
sns.histplot(final_leaderboard['Percentage'], kde=True)
plt.title('Distribution of Percentage Scores')
plt.xlabel('Percentage Score')
plt.ylabel('Frequency')
plt.show()


# Create a box plot of the total scores
plt.figure(figsize=(10, 6))
sns.boxplot(y='Total', data=final_leaderboard)
plt.title('Distribution of Total Scores')
plt.ylabel('Total Score')
plt.show()
```
